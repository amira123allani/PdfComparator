```python
import pdfminer
from pdfminer.layout import LAParams, LTTextBox
from pdfminer.pdfpage import PDFPage 
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.converter import PDFPageAggregator
rsrcmgr = PDFResourceManager()
laparams = LAParams()
device = PDFPageAggregator(rsrcmgr, laparams=laparams)
interpreter = PDFPageInterpreter(rsrcmgr, device)
def get_data (path):  
  fp = open(path, 'rb')
  rsrcmgr = PDFResourceManager()
  laparams = LAParams()
  device = PDFPageAggregator(rsrcmgr, laparams=laparams)
  interpreter = PDFPageInterpreter(rsrcmgr, device)
  pages = PDFPage.get_pages(fp)
 
  sent = {}
  loc ={}
  for pageNumber, page in enumerate(PDFPage.get_pages(fp)):

    interpreter.process_page(page)
    layout = device.get_result()
    location=[]
    l=[]
    for lobj in layout:
        if isinstance(lobj, LTTextBox):
            x0, y0_orig, x1, y1_orig  = lobj.bbox
            text = lobj.get_text()
            text.encode('utf-8')
            y0 = page.mediabox[3] - y1_orig
            y1 = page.mediabox[3] - y0_orig
            location.append([x0,y0,x1,y1])
            l.append(text)
    sent[pageNumber] = l
    loc[pageNumber] = location
  return sent , loc
```

```python
#pages = list(extract_pages('GS-98000-ED.pdf'))
from pdfrw import PdfReader, PdfWriter
def get_image_layout(layout_object):
    if isinstance(layout_object, pdfminer.layout.LTImage):
        return layout_object
    if isinstance(layout_object, pdfminer.layout.LTContainer):
        for child in layout_object:
           return get_image_layout(child)
    else:
        return None
   
def get_coor(layout,page):
    images = list(filter(bool, map(get_image_layout, layout)))
    l=[]
    for image in images :
        x0, y0_orig, x1, y1_orig  = image.bbox
        y0 = page.mediabox[3] - y1_orig
        y1 = page.mediabox[3] - y0_orig
        l.append((x0,y0,x1,y1))
    return l
  
def get_pdf_images(input_path , output_path) :
    doc = open(input_path,'rb')
    device = PDFPageAggregator(rsrcmgr, laparams=laparams)
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    images={}
    for pageNumber, page in enumerate(PDFPage.get_pages(doc)):
       interpreter.process_page(page)
       layout = device.get_result()
       l = get_coor(layout,page)
       if l != [] :
            images[pageNumber] = l
    x = PdfReader(input_path)
    output = PdfWriter()
    for key , value in images.items():
       for i in range(len(value)) :
          output.addpage(x.pages[key])
    output.write("new_file.pdf") 
    output = fitz.open("new_file.pdf")
    l= sum(list(images.values()),[])
    for page , box in zip(output,l) :
         page.set_cropbox(fitz.Rect(box))                        
    output.save(output_path)
    return images
```

```python
def sent_tokenization(data,location):
   for  key ,value in data.items() :
        sent = [sent_tokenize(txt) for txt in value]
        data[key]= sent
        
   for value , loc , key in zip(data.values() , location.values() , location.keys()):
            ll=[]
            for val , l in zip(value,loc) :
                l=[l]*len(val)
                ll.append(l) 
            location[key]=ll
    
   for key , val in data.items() :
        val = sum(list(val),[])
        data[key]= val
        
   for key , val in location.items() :
        val = sum(list(val),[])
        location[key]= val
           
   return data , location
```

```python
import math
import re
from collections import Counter

def get_cosine(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])

    sum1 = sum([vec1[x] ** 2 for x in list(vec1.keys())])
    sum2 = sum([vec2[x] ** 2 for x in list(vec2.keys())])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator

def text_to_vector(text):
    words = word_tokenize(text)
    from nltk.corpus import stopwords
    stopwords = set(stopwords.words(["english","german"]))
    l=[]
    for w in words:
      if (w.lower() not in stopwords) and (w not in string.punctuation):
        l.append(w)
    return Counter(l)

def cosine_sim(old_data , old_loc , new_data , new_loc):
   old_copy_data = old_data.copy()
   old_copy_loc = old_loc.copy()
   all_match = []  
   new_d=[]
   new_l=[]
   l_new_loc = list(new_loc.values())
   for key , value in new_data.items() :
    if key >= 2 :
      for new_sent , new_location in zip( value , l_new_loc[key] ) :
         l={}
         vector1 = text_to_vector(new_sent)
         five_pages_data = list(old_copy_data.items())[key-2 : key+3]
         five_pages_loc = list(old_copy_loc.items())[key-2 : key+3]
         for page_data ,page_loc in zip(five_pages_data,five_pages_loc):
            for old_sent,old_loc in zip(page_data[1],page_loc[1]) :  
                vector2 = text_to_vector(old_sent)
                cosine = get_cosine(vector1, vector2)
                l[cosine] = (page_data[0],old_loc,old_sent)
         if max(l.keys()) > 0.4 :
           match=[]
           indx_max = list(l).index(max(list(l)))
           match.extend((list(l.values())[indx_max],(key, new_location , new_sent)))
           all_match.append(match)
           old_copy_data[list(l.values())[indx_max][0]].remove(list(l.values())[indx_max][2])
           old_copy_loc[list(l.values())[indx_max][0]].remove(list(l.values())[indx_max][1])
         else :
           new_d.append([key,new_sent])
           new_l.append([key,new_location])
    else :
        for new_sent , new_location in zip( value , l_new_loc[key] ) :
         l={}
         vector1 = text_to_vector(new_sent)
         five_pages_data = list(old_copy_data.items())[0 : 6]
         five_pages_loc = list(old_copy_loc.items())[0 : 6]
         for page_data ,page_loc in zip(five_pages_data,five_pages_loc):
            for old_sent,old_loc in zip(page_data[1],page_loc[1]) :  
                vector2 = text_to_vector(old_sent)
                cosine = get_cosine(vector1, vector2)
                l[cosine] = (page_data[0],old_loc,old_sent)
         if max(l.keys()) > 0.4 :
           match=[]
           indx_max = list(l).index(max(list(l)))
           match.extend((list(l.values())[indx_max],(key, new_location , new_sent)))
           all_match.append(match)
           old_copy_data[list(l.values())[indx_max][0]].remove(list(l.values())[indx_max][2])
           old_copy_loc[list(l.values())[indx_max][0]].remove(list(l.values())[indx_max][1])
         else :
           new_d.append([key,new_sent])
           new_l.append([key,new_location])
        
   return all_match , old_copy_data , old_copy_loc, new_d , new_l
```

```python
def matching_fig(fig_list , n_images_loc , old_fig_list , o_images_loc):
    all_match = []  
    new_im = []
    new_im_l = []
    copy_old_fig = old_fig_list.copy()
    n_l=[]
    o_l=[]
    for key , value in n_images_loc.items():
        n_l.append([key]*len(value))
    n_pages= sum(n_l,[])
    n_location = sum(list(n_images_loc.values()),[])
    for key , value in o_images_loc.items():
        o_l.append([key]*len(value))
    o_pages= sum(o_l,[])
    o_location = sum(list(o_images_loc.values()),[])
    for fig , n_page, n_loc in zip(fig_list,n_pages,n_location):
       l={}
       orb = cv2.ORB_create(fastThreshold=0, edgeThreshold=0)
       kp_a, desc_a = orb.detectAndCompute(fig , None)
       for old_fig , o_page , o_loc in zip(copy_old_fig , o_pages, o_location ) :
              # detect keypoints and descriptors
              kp_b, desc_b = orb.detectAndCompute(old_fig , None)
              # define the bruteforce matcher object
              bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
            
              #perform matches.
              matches = bf.match(desc_a, desc_b)
              #Look for similar regions with distance < 50. Goes from 0 to 100 so pick a number between.
              similar_regions = [i for i in matches if i.distance < 50]  
              sim = len(similar_regions) / len(matches)
              l[sim] = (o_page , o_loc , old_fig)   
       if 0.4 <  max(l.keys()) < 0.95 :
           match=[]
           indx_max = list(l).index(max(list(l)))
           match.extend((list(l.values())[indx_max],(n_page , n_loc, fig)))
           all_match.append(match)
           copy_old_fig.remove(list(l.values())[indx_max][2])
           o_pages.remove(list(l.values())[indx_max][0])
           o_location.remove(list(l.values())[indx_max][1])
           #copy_old_loc[list(l.values())[indx_max][0]].remove(list(l.values())[indx_max][1])
       if  max(l.keys()) < 0.4 :
           new_im.append([n_loc[0],fig])
           new_im_l.append([n_loc[0],n_loc])
    location=[]
    for page , loc in zip(o_pages , o_location):
        location.append((page,loc))
        
    return all_match , copy_old_fig , location , new_im , new_im_l
```

```python
import difflib
import re

def tokenize(s):
    return re.split('\s+', s)
def untokenize(ts):
    return ' '.join(ts)
        
def equalize(s1, s2):
    l1 = tokenize(s1)
    l2 = tokenize(s2)
    res1 = []
    res2 = []
    prev = difflib.Match(0,0,0)
    for match in difflib.SequenceMatcher(a=l1, b=l2).get_matching_blocks():
        if (prev.a + prev.size != match.a):
            for i in range(prev.a + prev.size, match.a):
                res1 += [l1[i]]
            #res1 += l1[prev.a + prev.size:match.a]

        if (prev.b + prev.size != match.b):
            for i in range(prev.b + prev.size, match.b):
                res2 += [(l2[i])]
            #res2 += l2[prev.b + prev.size:match.b]
        prev = match
    return res1, res2

def compare(all_match) :
 n_modif_d=[]
 o_modif_d=[]
 n_modif_l=[]
 o_modif_l=[]
 for data in all_match :
    o_page , o_loc , o_data , n_page , n_loc , n_data = data[0][0],data[0][1],data[0][2],data[1][0],data[1][1],data[1][2]
    o_change , n_change = equalize(o_data,n_data)
    if o_change != [] :
        o_modif_d.append((o_page,o_change))
        o_modif_l.append((o_page,o_loc))
    if n_change != [] :
        n_modif_d.append((n_page,n_change))
        n_modif_l.append((n_page,n_loc))
   
 return n_modif_d , n_modif_l , o_modif_d ,o_modif_l
```

```python
import imutils
from skimage.metrics import structural_similarity
import cv2
import numpy as np
from PIL import Image

def compare_img(all_match):
  new_loc = []
  old_loc = []
  for data in all_match :
     old_im = data[0][2]
     new_im = data[1][2]
     x_shape = new_im.shape[1]
     y_shape = new_im.shape[0]
     x_org_shape = data[1][1][2] - data[1][1][0]
     y_org_shape = data[1][1][3] - data[1][1][1]
     o_copy = old_im.copy()
     n_copy = new_im.copy()
     old_im = cv2.resize(old_im, (new_im.shape[1],new_im.shape[0]))
     gray1 = cv2.cvtColor(old_im, cv2.COLOR_BGR2GRAY)
     gray2 = cv2.cvtColor(new_im, cv2.COLOR_BGR2GRAY)

     # Find the difference between the two images
     # Compute the mean structural similarity index between two images (similar).
     (similar, diff) = structural_similarity(gray1, gray2, full=True)
     # similar belongs in the interval [-1, 1] with 1 represents perfect similarity
     # diff is in range [0,1] so we need to convert it to an 8-bit array in range [0,255]
     diff = (diff*255).astype("uint8")

     # Apply threshold. Apply both THRESH_BINARY_INV and THRESH_OTSU
     thresh = cv2.threshold(diff, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]

     # Calculate contours 
     contours = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
     contours = imutils.grab_contours(contours)

     for contour in contours:
     # Calculate bounding box around contour
        if cv2.contourArea(contour) > 100 :
            x, y, w, h = cv2.boundingRect(contour)
            x0 = (x / x_shape)*x_org_shape
            y0 = (y / y_shape)*y_org_shape
            x1 = ((x_shape-(x+w))/x_shape)*x_org_shape
            y1 = ((y_shape-(y+h))/y_shape)*y_org_shape
            print(x0,y0,x1,y1)
            new_loc.append((data[0][0],[data[1][1][0]+x0 , data[1][1][1]+y0 , data[1][1][2]-x1 , data[1][1][3]-y1]))
            old_loc.append((data[0][0],[data[0][1][0]+x0 , data[0][1][1]+y0 , data[0][1][2]-x1 , data[0][1][3]-y1]))

  return old_loc , new_loc
```

```python
def sentence_similarity(sent1, sent2, stopwords=None):
    if stopwords is None:
        stopwords = []
 
    sent1 = [w.lower() for w in sent1]
    sent2 = [w.lower() for w in sent2]
 
    all_words = list(set(sent1 + sent2))
 
    vector1 = [0] * len(all_words)
    vector2 = [0] * len(all_words)
 
    # build the vector for the first sentence
    for w in sent1:
        if w in stopwords:
            continue
        vector1[all_words.index(w)] += 1
 
    # build the vector for the second sentence
    for w in sent2:
        if w in stopwords:
            continue
        vector2[all_words.index(w)] += 1
 
    return 1 - cosine_distance(vector1, vector2)
```

```python
def build_similarity_matrix(lower_case, stopwords=None):
    # Create an empty similarity matrix
    S = np.zeros([len(lower_case), len(lower_case)])
 
 
    for idx1 in range(len(lower_case)):
        for idx2 in range(len(lower_case)):
            if idx1 == idx2:
                continue
 
            S[idx1][idx2] = sentence_similarity(lower_case[idx1], lower_case[idx2], stopwords)
 
    # normalize the matrix row-wise
    for idx in range(len(S)):
        S[idx] /= S[idx].sum()
 
    return S
```

```python
def generate_summary(lower_case, top_n=5):
    # Remove all the stopwords in the document
    stop_words = stopwords.words('german')
    summarize_text = []
    
    
    
    #Read text and tokenize
    #lower_case  = nltk.word_tokenize(lower_case)
    
   
    
    #Generate similarity matrix across sentences
    sentence_similarity  = build_similarity_matrix((lower_case, stop_words))
    
    #Rank sentences in similarity matrix
    sentence_similiraty_graph = nx.from_numpy_array(sentence_similarity)
    scores = nx.pagerank(sentence_similiraty_graph , max_iter=10000000)
    
    #Sort the rank and pick top sentences
    ranked_sentence = sorted(((scores[i],s) for i,s in enumerate(lower_case)), reverse=True)    
    print("Indexes of top ranked_sentence order are ", ranked_sentence)
    
    for i in range(top_n):
        summarize_text.append(' '.join(ranked_sentence[i][1]))
        
    #Output the summarized text
    print('Summarized Text: \n', '. '.join(summarize_text))
```
